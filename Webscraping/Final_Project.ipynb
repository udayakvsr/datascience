```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Created my own methods and used in the code

#Extract data from the url
def extract_data(url):    
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text,'html.parser')
    return soup

#Dataset Information:    
def datainfo(df, info):
    print(info,'Information :')
    print('-----------------------')
    df.info()
    print('------------------------------------------------------------------------------------------------------------')
    print( info,'Data :')
    print('----------------')
    display(df)

#Preprocessing the data:
def preprocess(df):
    #Removing the duplicates:
    pp_df = df.drop_duplicates()
    #Dropping the null data: 
    pp_dff = pp_df.dropna()
    
    return pp_dff

#Web page title
def webtitle(html):
    return html.select('h1')[0].text
```

```python
# Took 5 webpages are from 5 different websites
webpages_1 = 'https://www.boxofficemojo.com/date/?ref_=bo_nb_rs_tab'
webpages_2 = 'https://www.nerdwallet.com/best/loans/personal-loans/personal-loan-interest-rates'
webpages_3 = 'https://washingtonmonthly.com/2022-college-guide/masters/'
webpages_4 = 'https://www.bankbazaar.com/gold-rate/gold-rate-trend-in-india.html'
webpages_5 = 'https://codepedia.info/twitter-trends/united-states'
```

```python
# Pulled data from the internet for Website-1:
website_1 = pd.read_html(webpages_1)
df_1 = website_1[0]
datainfo(df_1,'Website-1')
```

```python
# Pulled data from the internet for Website-2:
website_2 = pd.read_html(webpages_2)
df_2 = website_2[0]
datainfo(df_2,'Website-2')
```

```python
# Pulled data from the internet for Website-3:
website_3 = pd.read_html(webpages_3)
df_3 = website_3[0]
datainfo(df_3,'Website-3')
```

```python
# Pulled data from the internet for Website-4:
website_4 = pd.read_html(webpages_4)
df_4 = website_4[0]
datainfo(df_4,'Website-4')
```

```python
# Pulled data from the internet for Website-5:
website_5 = pd.read_html(webpages_5)
df_5 = website_5[0]
datainfo(df_5,'Website-5')
```

```python
# Preprocessing the data:
dff_1 = preprocess(df_1)
dff_2 = preprocess(df_2)
dff_3 = preprocess(df_3)
dff_4 = preprocess(df_4)
dff_5 = preprocess(df_5)
```

```python
#Describing the basic calculations on numerical data for Webpage-1:
print("Basic calculations on numerical data for Webpage-1:")
dff_1.describe()
```

```python
##Describing the basic calculations on numerical data for Webpage-2:
print("Basic calculations on numerical data for Webpage-2:")
dff_2['Min. credit score'] = dff_2['Min. credit score'].replace(['None'],'0')
dff_2['Min. credit score'] = pd.to_numeric(dff_2['Min. credit score'])
dff_2.describe()
```

```python
#Describing the basic calculations on numerical data for Webpage-3:
print("Basic calculations on numerical data for Webpage-3:")
dff_3.describe()
```

```python
#Describing the basic calculations on numerical data for Webpage-4:
print("Basic calculations on numerical data for Webpage-4:")
dff_4['Price (24 karat per 10 grams)'] = dff_4['Price (24 karat per 10 grams)'].replace('Rs.', ' ', regex=True)
dff_4['Price (24 karat per 10 grams)'] = dff_4['Price (24 karat per 10 grams)'].replace(',', '', regex=True)
dff_4['Price (24 karat per 10 grams)'] = pd.to_numeric(dff_4['Price (24 karat per 10 grams)'])
dff_4.describe()
```

```python
#Describing the basic calculations on numerical data for Webpage-5:
print("Basic calculations on numerical data for Webpage-5:")
dff_5.describe()
```

```python
# Produced chart for Webpage-1 and saving the figure::
Tot_Releases_Day = dff_1.groupby(['Day'], as_index=False)['Releases'].sum()

day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
ax = Tot_Releases_Day.set_index(Tot_Releases_Day.Day).loc[day_order].plot(kind="line", legend=False,figsize=(15,6))
ax.set_ylabel("Value")
plt.savefig("Total Movie Releases Per Day.png")
```

```python
# Produced chart for Webpage-2 and saving the figure:
dff_2.plot(x='Lender', y=['Min. credit score'], kind='bar',figsize=(15, 9),title = 'Min. Credit Score To Get A Loan From Lender')
plt.savefig("Min Credit Score To Get A Loan From Lender.png")
```

```python
# Produced chart for Webpage-3 and saving the figure::
dff3_out = dff_3.sort_values('Number of Pell graduates',ascending = False).groupby('Name')
dff_out = dff3_out.head(10)
plt.figure(figsize=(20,40))
plt.barh(dff_out['Name'].head(100), dff_out['Number of Pell graduates'].head(100))
plt.ylabel("University Name")
plt.xlabel("Total Number Of Pell Graduates")
plt.title("Total Number Of Pell Graduates For Top 100 Universities")
plt.show()
plt.savefig("Total Number Of Pell Graduates For Top 100 Universities.png")
```

```python
# Produced chart for Webpage-4 and saving the figure:
dff_4.plot(linestyle='dashed',x='Year',y='Price (24 karat per 10 grams)',figsize=(15, 9))
plt.savefig("Total Movie Releases Per Day.png")
```

```python
# Produced chart for Webpage-5 and saving the figure:
dff_5['Tweet Count'] = dff_5['Tweet Count'].replace(['below 10K'],'10000')
dff_5['Tweet Count'] = pd.to_numeric(dff_5['Tweet Count'])
dff_5.plot(x='Trending Hashtag / Topic', y=['Tweet Count'], kind='bar',figsize=(15, 9),title = 'Trending Hashtag Tweets')
plt.savefig("Trending Hashtag Tweets.png")
```

```python
# Saving the data to a file
dff_1.to_csv('webpages_1.csv', index=False)
dff_2.to_csv('webpages_2.csv', index=False)
dff_3.to_csv('webpages_3.csv', index=False)
dff_4.to_csv('webpages_4.csv', index=False)
dff_5.to_csv('webpages_5.csv', index=False)
```
